```python
import os, time, json, re
import pandas as pd
from tqdm.auto import tqdm
from dateutil import parser as dateparser
from apify_client import ApifyClient
from google.colab import userdata
```

```python

APIFY_TOKEN = userdata.get('APIFY_API_KEY')
INPUT_CSV   = "/content/mlops-events.cleaned.csv"
OUTPUT_CSV  = "/content/mlops-events-enriched.csv"
CACHE_PATH  = "/content/apify_youtube_cache.json"  # resume cache


YOUTUBE_URL_COL = "YouTube Link"

APIFY_ACTOR_ID  = "CTQcdDtqW5dvELvur"

SLEEP_SEC_BETWEEN_CALLS = 1.0
MAX_RETRIES = 3
CHECKPOINT_EVERY = 20    # save partial CSV every N processed rows
```

```python
client = ApifyClient(APIFY_TOKEN)
```

```python
def load_json_cache(path: str) -> dict:
    """Load {url: {yt_views, yt_published_date, yt_transcript}}"""
    if os.path.exists(path):
        with open(path, "r", encoding="utf-8") as f:
            try:
                return json.load(f)
            except Exception:
                return {}
    return {}

def save_json_cache(path: str, cache: dict) -> None:
    tmp = path + ".tmp"
    with open(tmp, "w", encoding="utf-8") as f:
        json.dump(cache, f, ensure_ascii=False)
    os.replace(tmp, path)

def parse_views_to_int(views_str: str):
    """Converts '23 views', '1,234 views', '12K views', etc. to int."""
    if not views_str or not isinstance(views_str, str):
        return None
    s = views_str.lower().strip()
    s = s.replace(" views", "").replace("view", "").strip()
    m = re.match(r"^([\d,\.]+)\s*([kmb])?$", s)
    if not m:
        digits = re.sub(r"[^\d]", "", s)
        return int(digits) if digits.isdigit() else None
    num, suffix = m.groups()
    num = float(num.replace(",", ""))
    mult = {"k": 1_000, "m": 1_000_000, "b": 1_000_000_000}.get(suffix, 1)
    return int(num * mult)
```

```python
def call_apify_with_retries(url: str):
    """Run Apify actor for a single URL with retries + backoff."""
    run_input = {
        "includeTimestamps": "Yes",
        "language": "English",
        "startUrls": [url],
    }
    delay = 1.0
    last_exc = None
    for attempt in range(1, MAX_RETRIES + 1):
        try:
            run = client.actor(APIFY_ACTOR_ID).call(run_input=run_input)
            dataset = client.dataset(run["defaultDatasetId"])
            # Typically one item per URL
            for item in dataset.iterate_items():
                views_raw = item.get("views")
                published = item.get("videoPostDate")
                transcript = item.get("transcript")  # list[{text, timestamp}]
                return {
                    "yt_views": parse_views_to_int(views_raw),
                    "yt_published_date": published,
                    "yt_transcript": transcript,  # keep EXACT as returned
                }
            # No items returned
            return {}
        except Exception as e:
            last_exc = e
            if attempt < MAX_RETRIES:
                time.sleep(delay)
                delay *= 2
            else:
                raise e
    # Should not reach here
    raise last_exc

```

```python
# ==== Load CSV (support resuming from OUTPUT_CSV if it exists) ====
if os.path.exists(OUTPUT_CSV):
    df = pd.read_csv(OUTPUT_CSV)
else:
    df = pd.read_csv(INPUT_CSV)

# Ensure required columns exist
for col in ["yt_views", "yt_published_date", "yt_transcript"]:
    if col not in df.columns:
        df[col] = None

# Normalize URL column if present
if YOUTUBE_URL_COL not in df.columns:
    raise ValueError(f"Input CSV must contain column '{YOUTUBE_URL_COL}'")

# Load or initialize cache
cache = load_json_cache(CACHE_PATH)

# Helper: check if a row seems already enriched
def row_has_transcript(val) -> bool:
    if pd.isna(val):
        return False
    s = str(val).strip()
    return s not in ["", "None", "nan"]
```

```python
# ==== Process rows (resume-safe) ====
failures = []
processed_since_checkpoint = 0

for idx in tqdm(range(len(df)), desc="Enriching"):
    url = str(df.loc[idx, YOUTUBE_URL_COL]).strip()
    if not url or url.lower() == "nan":
        continue

    # Short-circuit if already present in dataframe
    if row_has_transcript(df.loc[idx, "yt_transcript"]):
        continue

    # Use cache if available
    if url in cache:
        data = cache[url]
    else:
        try:
            data = call_apify_with_retries(url)
            # Even empty dict is cached, so we don't retry needlessly next run
            cache[url] = data
            save_json_cache(CACHE_PATH, cache)
            time.sleep(SLEEP_SEC_BETWEEN_CALLS)
        except Exception as e:
            failures.append((idx, url, repr(e)))
            # Cache the failure as empty dict to avoid tight-loop retries on resume
            cache[url] = {}
            save_json_cache(CACHE_PATH, cache)
            continue

    if data:
        df.at[idx, "yt_views"] = data.get("yt_views")
        df.at[idx, "yt_published_date"] = data.get("yt_published_date")
        # Store transcript EXACTLY as returned, JSON-encoded in CSV
        df.at[idx, "yt_transcript"] = json.dumps(data.get("yt_transcript"), ensure_ascii=False)
    else:
        # No usable data returned
        df.at[idx, "yt_views"] = df.at[idx, "yt_views"] if pd.notna(df.at[idx, "yt_views"]) else None
        df.at[idx, "yt_published_date"] = df.at[idx, "yt_published_date"] if pd.notna(df.at[idx, "yt_published_date"]) else None
        df.at[idx, "yt_transcript"] = df.at[idx, "yt_transcript"] if row_has_transcript(df.at[idx, "yt_transcript"]) else None

    processed_since_checkpoint += 1
    if processed_since_checkpoint >= CHECKPOINT_EVERY:
        df.to_csv(OUTPUT_CSV, index=False)
        processed_since_checkpoint = 0
```
